**Checkpoint 1: Data Cleaning 1**

```python
#code for importing the companies.txt file here
companies = pd.read_csv("Companies.txt", delimiter ='\t')
```

```python
companies
```

```python
#decoding the company dataframe value
companies.permalink=companies.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
companies.name=companies.name.str.encode('utf-8').str.decode('ascii', 'ignore')

#making all the values in the permalink companies datafram to lowercase
companies.permalink = companies.permalink.str.lower()
```

```python
companies.head()
```

```python
#code for importing the rounds2.csv file here
rounds2 = pd.read_csv("rounds2.csv", encoding = "ISO-8859-1")
```

```python
rounds2
```

```python
#decoding the rounds2 dataframe value
rounds2.company_permalink=rounds2.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')

#making all the values in the company_permalink in rounds2 dataframe to lowercase
rounds2.company_permalink = rounds2.company_permalink.str.lower()
```

```python
rounds2.head()
```

```python
#unique companies in in rounds2
rounds2['company_permalink'].nunique()
```

```python
#unique companies in in companies
companies['permalink'].nunique()
```

```python
#companies in the rounds2 file which are not  present in companies 
companies[~companies['permalink'].isin(rounds2['company_permalink'])]
```

```python
#Merge the two data frames so that all variables (columns) in the companies frame are added to the rounds2 data frame.
master_frame = pd.merge(rounds2, companies, how='left', left_on='company_permalink',right_on='permalink')
```

```python
master_frame.head()
```

```python
#shape of rounds2 df
rounds2.shape
```

```python
#shape of master_frame df
master_frame.shape
```

```python
master_frame.info()
```

```python
master_frame
```

```python
#finding the sum of null values in a column
master_frame.isnull().sum()
```

**Cleaning the data**

```python
## dropping rows where raised_amount_usd value is null
master_frame=master_frame[~np.isnan(master_frame['raised_amount_usd'])]
```

```python
## dropping rows where category_list value is null
master_frame = master_frame.dropna(subset=['category_list']) 
```

```python
## dropping rows where country_code value is null
master_frame = master_frame.dropna(subset=['country_code']) 
```

```python
master_frame.isnull().sum()
```

**Checkpoint 4: Sector Analysis 1**

```python
#importing the mapping.csv file here
mapping = pd.read_csv("mapping.csv")
```

```python
#replacing the irrelevant '0' by 'na' in the category_list in mapping frame
mapping['category_list'] = np.where(mapping['category_list']=='Enterprise 2.0', 'Enterprise 2.0', 
                                   mapping['category_list'].str.replace('0','na'))
```

```python
mapping.head()
```

```python
#category_df dataframe - seperating the sub-sectors separated by a pipe as seperate columns and fetching the primary sector
category_df = master_frame["category_list"].str.split("|", n = -1, expand = True)
```

```python
#adding the primary sector to the master_frame df
master_frame['primary_sector'] = category_df[0]
```

```python
master_frame
```

```python
#mapping the primary sector values to main sector by using melt function
mapping_main=pd.DataFrame()
mapping_main[['primary_sector','main_sector','value']] = pd.melt(mapping, id_vars = ['category_list'], var_name = ['Main_sector'])
mapping_main = mapping_main[mapping_main.value !=0 ]
```

```python
mapping_main.head()
```

```python
#adding the main sector to the master_frame df
master_frame=pd.merge(master_frame,mapping_main[['primary_sector','main_sector']],on='primary_sector',how='left')
```

```python
master_frame
```

```python
#exporting the 'master_frame.xlsx' file to local directory to import the same to tableau 
master_frame.to_excel('master_frame.xlsx')
```

```python
#modifiying sectorframe dataframe with only funding round type "venture" with modified master_frame
Sector_frame = master_frame.loc[(master_frame['funding_round_type'] == 'venture')]  
```

```python
Sector_frame
```

**Checkpoint 5: Sector Analysis 2**

```python
#Create three separate data frames D1 for country USA
#All the columns of the master_frame along with the primary sector and the main sector, investments range of 5-15 M USD
D1 = Sector_frame.loc[(Sector_frame['country_code'] == 'USA') & (Sector_frame['raised_amount_usd'] >= 5) & 
                    (Sector_frame['raised_amount_usd'] < 15)]
```

```python
D1
```

```python
#grouping D1 dataframe by main sector 
D1_by_mainsec = D1.groupby('main_sector')
```

```python
#D1_investment_count dataframe with main sector wise total count  
D1_investment = pd.DataFrame()
D1_investment['total_investment_count'] = D1_by_mainsec['raised_amount_usd'].count()

#D1_investment_sum dataframe with main sector wise total sum  
D1_investment['total_investment_sum'] = D1_by_mainsec['raised_amount_usd'].sum()
```

```python
D1_investment.sort_values('total_investment_count',ascending=False)
```

```python
#Total number of Investments (count)
D1_investment.total_investment_count.sum()
```

```python
#Total amount of investment (USD)
D1_investment.total_investment_sum.sum()
```

```python
#The total number (or count) of investments for each main sector in a separate column
D1 = pd.merge(D1, D1_investment[['total_investment_count']], how = 'left', on = 'main_sector')
#The total amount invested in each main sector in a separate column
D1 = pd.merge(D1, D1_investment[['total_investment_sum']], how = 'left', on = 'main_sector')
```

```python
D1
```

```python
#For the top sector which is Others, sorting the raised_amount_usd to find thecompany received the 
#highest investment
D1_top_sector = D1.loc[(D1.main_sector=='Others')].sort_values('raised_amount_usd',ascending=False)
D1_top=D1_top_sector.groupby('name')
D1_top['raised_amount_usd'].sum().to_frame().sort_values('raised_amount_usd',ascending=False)
```

```python
#For the second-best sector which is Social, Finance, Analytics, Advertising, sorting the raised_amount_usd to find thecompany received the 
#highest investment
D1_second_sector=D1.loc[(D1.main_sector=='Social, Finance, Analytics, Advertising')].sort_values('raised_amount_usd',ascending=False)
D1_second=D1_second_sector.groupby('name')
D1_second['raised_amount_usd'].sum().to_frame().sort_values('raised_amount_usd',ascending=False)
```
